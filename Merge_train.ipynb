## ANN

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

from keras.models import Sequential
from keras.layers import Dense, Activation, Embedding, Flatten, LeakyReLU, BatchNormalization,Dropout,BatchNormalization
from keras.activations import relu, sigmoid
from keras.optimizers import SGD,schedules
```

```python
def create_model(layers,activation,dropout,initializer,optimizer):
  model = Sequential()
  for i,nodes in enumerate(layers):
    if i==0:
      model.add(Dense(nodes,input_dim=x_train.shape[1]))
      model.add(Activation(activation))
      model.add(Dropout(dropout))
      model.add(BatchNormalization(axis=-1,momentum=0.99,epsilon=0.001))
    else:
      model.add(Dense(nodes))
      model.add(Activation(activation))
      model.add(Dropout(dropout))
      model.add(BatchNormalization(axis=-1,momentum=0.99,epsilon=0.001))
  
    
  model.add(Dense(units=1,kernel_initializer=initializer,activation='sigmoid' ))
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  
  return model
```

```python
model = KerasClassifier(build_fn=create_model, verbose=2)
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)


layers = [(160,100,60)]
activations = ['relu']
dropout = [0.1]
initializer = ['glorot_uniform']
optimizer= ['sgd']
#initial = ['random_normal','truncated_normal','random_uniform','zeros','ones','glorot_normal','glorot_uniform',
#           'he_normal','he_uniform','identity','orthogonal','constant','variance_scaling']

params_grid = dict(layers=layers, activation=activations,dropout=dropout,initializer=initializer,
                   optimizer= optimizer,batch_size=[32], epochs=[256])

grid = GridSearchCV(estimator=model, param_grid=params_grid, cv=skf,verbose=10,)

 
grid_result = grid.fit(x_train,y_train)
```
